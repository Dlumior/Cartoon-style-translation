```python
import cv2
def isgray(imgpath):
    img = cv2.imread(imgpath)
    if len(img.shape) < 3: return True
    if img.shape[2]  == 1: return True
    return False
```

```python
def delete_corrupt_files(path, format=".jpg"):
  for i, filename in enumerate(listdir(path)):
    #print(filename)
    filename = join(path, filename)
    if filename.endswith((format, ".jpeg")):
      if isgray(filename):
        remove(filename)
        print("Removiendo archivo no RGB:", filename)
        continue
      if any(word in filename for word in ["gif", "png", "webp"]):
        remove(filename)
        print("Removiendo archivo no JPG:", filename)
        continue

      try:
        img = Image.open(filename) #Abrir la imagen
        img.verify() #Verificar que sea una imagen
        rename(filename, f'{path}/{i}.jpeg')
      except:
        remove(filename)
        print("Removiendo archivo corrupto:", filename)
    else:
      remove(filename)
      print("Removiendo archivo no jpg:", filename)
```

```python
delete_corrupt_files("download_images_1/adventure_time")
```

```python
delete_corrupt_files("download_images_1/landscape_draw")
```

```python
import os

def get_original_filename(filename):
  return filename.split('.', maxsplit=1).pop()

def delete_repeated(path):
  filenames = listdir(path)
  no_repeated = {get_original_filename(filename): filename for filename in filenames}

  for filename in filenames:
    if filename == no_repeated.get(get_original_filename(filename)):
      print('saving   ===>', filename)
      continue
    print('deleting ===>', filename)
    os.remove(f'{path}/{filename}')
```

```python
delete_repeated("download_images/adventure_time")
```

# Cargar las imagenes

```python
from os.path import isfile
```

```python
path = "download_images"
```

```python
car_path = path + "/landscape_draw"

car_files = list()

for filename in listdir(car_path):
  filename = join(car_path, filename)
  if isfile(filename):
    car_files.append(filename)
```

```python
truck_path = path + "/adventure_time"

truck_files = list()

for filename in listdir(truck_path):
  filename = join(truck_path, filename)
  if isfile(filename):
    truck_files.append(filename)
```

```python
from random import choice
import matplotlib.pyplot as plt

def plot_random_image(list_of_images_paths):
  #Seleccionar aleatoriamente una imagen
  random_image_path = choice(list_of_images_paths)

  #Cargar la imagen
  image = Image.open(random_image_path).convert("RGBA")

  #Mostrar la imagen
  plt.imshow(image)
  plt.axis("off")
```

```python
plot_random_image(car_files)
```

```python
plot_random_image(truck_files)
```
